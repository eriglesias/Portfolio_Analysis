```python
#Import data from Yahoo Finance

import pandas_datareader.data as web
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

```

<h2> Import Data </h2>

```python


start = dt.datetime(2012,1,1)
end = dt.datetime(2022,8,1)
stocks = ['TSLA','GM','F','DIS','NFLX','LHA.DE','INTC','LMT','JBHT','MS']
#daily = web.DataReader('TSLA','GM', start, end )
daily = web.get_data_yahoo(stocks,start,end)
daily.head()
```

```python
Get Close Price
```

```python
close = daily['Close']
ax = close.plot(title='Stocks')
ax.set_xlabel('Date')
ax.set_ylabel('Close')
ax.grid()
plt.show()
```
